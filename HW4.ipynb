```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# constants
T_1_5 = 31.29 # degC at z1=1.5m
T_3_0 = 30.87 # degC at z3=3.0m
u_2_0 = 3.0   # m/s  at z2=2.0m
z_0 = 0.01    # surface roughness
d_0 = 0.0     # displacement
k = 0.4       # von Karman constant
rho = 1.16    # kg/m^3 density of water
cp = 1005     # J kg^-2 K^-1 heat capacity of air
Le = 2.3e6    # latent heat of vaporization
Ch_ = 1e-3     # Stanton number
Ch = (k)/((np.log(3))*(np.log(1.5)))

# functions
def Obukhov_length(USTAR, TA, TS, K, H, RHO, CP):
    return (-USTAR**3)/((K*9.8/TA)*(H/(RHO*CP)))

def Psi(zeta): # equation 2.58
    if zeta > 1:
        return -5*zeta - 5*np.log(zeta)
    else:
        return -5*zeta

def USTAR(K, U, Z, Z0, D0, L):
    return U*K*(np.log((Z-D0)/Z0)-Psi((Z-D0)/L) + Psi(Z0/L))**(-1)

def H_heat(K, USTAR, Z, Z0, D0, L, TA, TS, RHO, CP):
    return (TS-TA)*K*USTAR*RHO*CP*(np.log((Z-D0)/Z0)-Psi((Z-D0)/L) + Psi(Z0/L))**(-1)
```

```python
# part a
# initial conditions before the iteration 
u_star = u_2_0*k*(np.log((2-d_0)/z_0))**(-1) # initial u* from equation 2.41
H = Ch*rho*cp*u_2_0*(T_1_5 - T_3_0) # initial H using equation 2.38
L = Obukhov_length(u_star, T_3_0, T_1_5, k, H, rho,cp)

# iterate 10 times 
for i in range(9):
    u_star = USTAR(k,u_2_0,2,z_0,d_0,L)
    H = H_heat(k, u_star, 2, z_0, d_0, L, T_3_0, T_1_5, rho, cp)
    L = Obukhov_length(u_star, T_3_0, T_1_5, k, H, rho,cp)
```

```python
print(u_star)
print(H)
print(L)
```

```python
# part b (LeE = H - Rn)

# constants
Le = 2.3e6    # latent heat of vaporization
Rn = 392      # W/m^2 net radiation

# net rate of evaporation in W/m^2 (LeE = H - Rn)
E_W_m_2 = Rn - H
print(E_W_m_2)

# net rate of evaporation in mm day-1 (E = (H - Rn)/Le)
E_m_s = E_W_m_2/Le
E_mm_month = E_m_s*86400*1000/31
print(E_mm_month)
```
