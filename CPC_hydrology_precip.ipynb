```python
# import packages
%pylab inline

import xarray as xr
import pandas as pd

import os

import statsmodels.api as sm
import statsmodels.formula.api as smf

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from mpl_toolkits.axes_grid1 import make_axes_locatable
from cartopy.util import add_cyclic_point
```

```python
# create xarray for daya
dloc = "/projects/GEOCLIM/el2358/projects/cpc_data"
D = xr.open_mfdataset("{}/precip.*.nc".format(dloc))
```

```python
# data
D = D['precip'].load()
```

```python
# getting daily precipitation in our water year in the region of interest
D_watershed = D.sel(lat=slice(34, 33)).sel(lon=slice(360-79.6,360-78)).sel(time=slice("2018-10-01", "2019-09-30"))

# average spatially
D_watershed_mean = D_watershed.mean('lat').mean('lon')

# calculate total precip in the Water Year
precip_total = D_watershed_mean.sum('time')
precip_total
```
